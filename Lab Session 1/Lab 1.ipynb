This is a small example of how to wrangle data.

```python
import pandas as pd
import matplotlib as plt

%matplotlib inline
```

```python
file1 = pd.read_csv('file1.csv', header=None, names=['a', 'b'])
file1.head()
```

```python
file1.plot.scatter(x='a', y='b')
```

```python
file2 = pd.read_csv('file2.csv', header=None)
file2.head()
```

```python
left = file2.iloc[:,[0,1]]
right = file2.iloc[:,[2,3]]
left
```

```python
left = left.rename(columns={0:'a', 1:'b'})
left
```

```python
right = right.rename(columns={2:'a', 3:'b'})
right
```

```python
file2 = pd.concat([left,right])
file2
```

```python
file2.plot.scatter(x='a', y='b')
```

```python
file3 = pd.read_csv('file3.csv', header=None)
file3.head()
```

```python
file3 = file3.T
```

```python
file3.rename(columns={0:'a',1:'b'}, inplace=True)
file3
```

```python
file3.plot.scatter(x='a', y='b')
```

```python
file1['c'] = file1['a']
file2['c'] = file2['a']+100
file3['c'] = file3['a']+180
```

```python
file_final = pd.concat([file1,file2,file3])
```

```python
file_final
```

```python
file_final.plot.scatter(x='c', y='b')
```

```python
file_final.to_csv('final.csv', sep=';')
```
